```python
hapdf = pd.read_csv('2016.csv')
hapdf.head()
```

```python
cols_to_drop = ['Region','Happiness Rank','Lower Confidence Interval','Upper Confidence Interval',\
                'Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom',\
                'Generosity','Dystopia Residual','Trust (Government Corruption)']

hapdf.drop(cols_to_drop, 1, inplace=True)
```

```python
bigmacdf = pd.read_csv('big-mac-source-data.csv')
bigmacdf['date'] = pd.to_datetime(bigmacdf['date']) 
```

```python
mask = (bigmacdf['date'] >= '2016-01-01') & (bigmacdf['date'] <= '2016-12-31')
bigmacdf = bigmacdf.loc[mask]
```

```python
cols_to_drop = ['GDP_dollar']
bigmacdf.drop(cols_to_drop, axis=1, inplace=True)
```
