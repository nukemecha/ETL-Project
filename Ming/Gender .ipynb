```python
import pandas as pd
```

```python
gender_df=pd.read_csv("Gender_StatsData.csv")
gender_df.head()
gender_df_2016=gender_df.loc[:,["Country Code","Indicator Name","2016"]]
```

```python
gender_clean=gender_df_2016.dropna()
gender_clean
```

```python
total_population=gender_clean.loc[gender_clean["Indicator Name"]=="Population, total" ]
total_population
total_clean=total_population.drop(columns="Indicator Name")
total_clean.rename(columns ={'Country Code':'iso_a3',"2016":"Total Population"}, inplace = True)
total_clean
```

```python
female_population=gender_clean.loc[gender_clean["Indicator Name"]=="Population, female" ]
female_population
female_clean=female_population.drop(columns="Indicator Name")
female_clean.rename(columns ={'Country Code':'iso_a3',"2016":"Female Population"}, inplace = True)
female_clean
```

```python
gender_merged=pd.merge(female_clean,total_clean,on="iso_a3", how="left")
```

```python
gender_merged
```

```python
gender_merged["Female/Total"]=gender_merged["Female Population"]/gender_merged["Total Population"]
```

```python
gender_merged
```

```python
gender_merged.to_csv("gender data.csv")
```
